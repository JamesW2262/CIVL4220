```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.callbacks import History
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import TimeSeriesSplit
import itertools
import os
from matplotlib.dates import YearLocator, DateFormatter

# Load and prepare data
df = pd.read_csv('electricity consumption.csv')
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str))
df.set_index('Date', inplace=True)
numeric_cols = ['Domestic', 'Commercial', 'Industrial', 'Street lighting (2)', 'All groups']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Anomaly detection and handling
def detect_and_handle_anomalies(series, threshold=3):
    z_scores = (series - series.mean()) / series.std()
    outliers = np.abs(z_scores) > threshold
    series[outliers] = np.nan
    series = series.interpolate(method='linear')  # Interpolate missing values
    return series

# Apply anomaly handling to each column
for col in numeric_cols:
    df[col] = detect_and_handle_anomalies(df[col])

df = df.dropna()

# Feature engineering
def add_features(df):
    df = df.copy()
    for col in numeric_cols:
        df[f'{col}_lag_1'] = df[col].shift(1)
        df[f'{col}_lag_12'] = df[col].shift(12)
        df[f'{col}_rolling_mean_12'] = df[col].rolling(window=12).mean()
    df['Month'] = df.index.month
    df['Year'] = df.index.year
    df['Season'] = df['Month'].map({1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring',
                                    6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall',
                                    11: 'Fall', 12: 'Winter'})
    df = pd.get_dummies(df, columns=['Season'], drop_first=True)
    return df.dropna()

df = add_features(df)

# Configure plotting style
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.grid'] = True

# Error metrics calculation
def calculate_metrics(actual, predicted):
    actual, predicted = np.array(actual), np.array(predicted)
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100 if np.all(actual != 0) else np.nan
    return {'MAE': mae, 'RMSE': rmse, 'MAPE': mape}

# Time-series cross-validation for SARIMA
def sarima_cross_validation(series, order=(1,1,1), seasonal_order=(1,1,1,12), n_splits=5):
    tscv = TimeSeriesSplit(n_splits=n_splits)
    metrics = {'MAE': [], 'RMSE': [], 'MAPE': []}

    for train_index, test_index in tscv.split(series):
        train, test = series.iloc[train_index], series.iloc[test_index]
        try:
            model = SARIMAX(train, order=order, seasonal_order=seasonal_order,
                           enforce_stationarity=False, enforce_invertibility=False)
            model_fit = model.fit(disp=False)
            forecast = model_fit.forecast(steps=len(test))
            metric = calculate_metrics(test, forecast)
            metrics['MAE'].append(metric['MAE'])
            metrics['RMSE'].append(metric['RMSE'])
            metrics['MAPE'].append(metric['MAPE'])
        except Exception as e:
            print(f"SARIMA CV failed for {series.name}: {str(e)}")
            metrics['MAE'].append(np.nan)
            metrics['RMSE'].append(np.nan)
            metrics['MAPE'].append(np.nan)

    return {k: np.nanmean(v) for k, v in metrics.items()}

# Time-series cross-validation for LSTM
def lstm_cross_validation(series, n_steps=12, n_splits=5):
    tscv = TimeSeriesSplit(n_splits=n_splits)
    metrics = {'MAE': [], 'RMSE': [], 'MAPE': []}
    scaler = MinMaxScaler(feature_range=(0, 1))

    for train_index, test_index in tscv.split(series):
        train, test = series.iloc[train_index], series.iloc[test_index]
        scaled_train = scaler.fit_transform(train.values.reshape(-1, 1))

        X, y = [], []
        for i in range(len(scaled_train) - n_steps):
            X.append(scaled_train[i:(i+n_steps), 0])
            y.append(scaled_train[i+n_steps, 0])
        if len(X) == 0:
            continue
        X, y = np.array(X), np.array(y)
        X = X.reshape(X.shape[0], X.shape[1], 1)

        model = Sequential()
        model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')
        model.fit(X, y, epochs=50, verbose=0, batch_size=32)

        test_inputs = scaler.transform(test.values.reshape(-1, 1))
        X_test = []
        for i in range(len(test_inputs) - n_steps):
            X_test.append(test_inputs[i:(i+n_steps), 0])
        if not X_test:
            continue
        X_test = np.array(X_test).reshape(-1, n_steps, 1)
        predictions = model.predict(X_test, verbose=0)
        predictions = scaler.inverse_transform(predictions).flatten()
        actual = test.values[n_steps:]

        metric = calculate_metrics(actual, predictions[:len(actual)])
        metrics['MAE'].append(metric['MAE'])
        metrics['RMSE'].append(metric['RMSE'])
        metrics['MAPE'].append(metric['MAPE'])

    return {k: np.nanmean(v) for k, v in metrics.items()}

# Hyperparameter tuning for SARIMA
def tune_sarima(series, p_range=(0,2), d_range=(0,2), q_range=(0,2), s=12):
    best_aic = float('inf')
    best_order = None
    best_seasonal_order = None

    for p, d, q in itertools.product(range(*p_range), range(*d_range), range(*q_range)):
        for P, D, Q in itertools.product(range(*p_range), range(*d_range), range(*q_range)):
            try:
                model = SARIMAX(series, order=(p,d,q), seasonal_order=(P,D,Q,s),
                               enforce_stationarity=False, enforce_invertibility=False)
                model_fit = model.fit(disp=False)
                if model_fit.aic < best_aic:
                    best_aic = model_fit.aic
                    best_order = (p,d,q)
                    best_seasonal_order = (P,D,Q,s)
            except:
                continue

    print(f"Best SARIMA parameters for {series.name}: order={best_order}, seasonal_order={best_seasonal_order}")
    return best_order, best_seasonal_order

# Enhanced SARIMA forecasting
def sarima_forecast(series, steps, order=(1,1,1), seasonal_order=(1,1,1,12)):
    try:
        model = SARIMAX(series, order=order, seasonal_order=seasonal_order,
                       enforce_stationarity=False, enforce_invertibility=False)
        model_fit = model.fit(disp=False)
        forecast = model_fit.get_forecast(steps=steps)
        return forecast.predicted_mean, model_fit
    except Exception as e:
        print(f"SARIMA failed for {series.name}: {str(e)}")
        return pd.Series([np.nan]*steps, index=pd.date_range(series.index[-1] + pd.DateOffset(months=1),
                                                            periods=steps, freq='MS')), None

# Enhanced LSTM forecasting with feature engineering
def lstm_forecast(series, df_features, steps, n_steps=12):
    try:
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(series.values.reshape(-1, 1))

        # Include additional features
        feature_cols = [f'{series.name}_lag_1', f'{series.name}_lag_12', f'{series.name}_rolling_mean_12',
                        'Season_Spring', 'Season_Summer', 'Season_Winter']
        features = df_features[feature_cols].values
        scaled_features = scaler.fit_transform(features)

        X, y = [], []
        for i in range(len(scaled_data) - n_steps):
            X.append(np.hstack((scaled_data[i:(i+n_steps), 0].reshape(-1, 1),
                                scaled_features[i:(i+n_steps)])))
            y.append(scaled_data[i+n_steps, 0])
        X, y = np.array(X), np.array(y)
        X = X.reshape(X.shape[0], X.shape[1], X.shape[2])

        model = Sequential()
        model.add(LSTM(150, activation='relu', input_shape=(n_steps, X.shape[2])))
        model.add(LSTM(150, return_sequences=True))
        model.add(Dropout(0.3))
        model.add(LSTM(150))
        model.add(Dropout(0.3))

        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')

        history = History()
        model.fit(X, y, epochs=200, verbose=0, batch_size=32, callbacks=[history])

        # Forecast
        test_inputs = scaled_data[-n_steps:].reshape(1, n_steps, 1)
        test_features = scaled_features[-n_steps:].reshape(1, n_steps, -1)
        lstm_forecast = []
        last_date = series.index[-1]

        for _ in range(steps):
            x_input = np.hstack((test_inputs[:, -n_steps:, :], test_features[:, -n_steps:, :]))
            yhat = model.predict(x_input, verbose=0)[0][0]
            lstm_forecast.append(yhat)

            # Update inputs with forecasted value
            test_inputs = np.append(test_inputs[:, 1:, :], [[[yhat]]], axis=1)

            # Update features (simple extrapolation for lags and rolling mean)
            new_lag_1 = yhat
            new_lag_12 = scaled_data[-12] if len(scaled_data) >= 12 else yhat
            new_rolling_mean = np.mean(scaled_data[-12:]) if len(scaled_data) >= 12 else yhat
            new_season = df_features.iloc[-1][['Season_Spring', 'Season_Summer', 'Season_Winter']].values
            new_features = scaler.transform([[new_lag_1, new_lag_12, new_rolling_mean] + list(new_season)])
            test_features = np.append(test_features[:, 1:, :], [new_features], axis=1)

        lstm_forecast = scaler.inverse_transform(np.array(lstm_forecast).reshape(-1, 1)).flatten()
        dates = pd.date_range(last_date + pd.DateOffset(months=1), periods=steps, freq='MS')
        return pd.Series(lstm_forecast, index=dates), history
    except Exception as e:
        print(f"LSTM failed for {series.name}: {str(e)}")
        return pd.Series([np.nan]*steps, index=pd.date_range(series.index[-1] + pd.DateOffset(months=1),
                                                            periods=steps, freq='MS')), None

# Main forecasting loop with cross-validation
forecast_years = 10
forecast_months = forecast_years * 12
results = {}
cv_results = {}

for group in numeric_cols:
    print(f"\n{'='*50}")
    print(f"Processing {group}...")
    print(f"{'='*50}")

    series = df[group].dropna()

    # Perform cross-validation
    print(f"Running SARIMA cross-validation for {group}...")
    sarima_cv_metrics = sarima_cross_validation(series)
    print(f"SARIMA CV Metrics: {sarima_cv_metrics}")

    print(f"Running LSTM cross-validation for {group}...")
    lstm_cv_metrics = lstm_cross_validation(series)
    print(f"LSTM CV Metrics: {lstm_cv_metrics}")

    cv_results[group] = {'SARIMA': sarima_cv_metrics, 'LSTM': lstm_cv_metrics}

    # Tune SARIMA parameters
    print(f"Tuning SARIMA parameters for {group}...")
    best_order, best_seasonal_order = tune_sarima(series, p_range=(0,2), d_range=(0,2), q_range=(0,2))

    # SARIMA forecast with tuned parameters
    sarima_fc, sarima_model = sarima_forecast(series, forecast_months, best_order, best_seasonal_order)

    # LSTM forecast with features
    lstm_fc, lstm_history = lstm_forecast(series, df, forecast_months)

    # Dynamic hybrid weighting based on CV performance
    sarima_mape = cv_results[group]['SARIMA']['MAPE']
    lstm_mape = cv_results[group]['LSTM']['MAPE']
    total_mape = sarima_mape + lstm_mape
    sarima_weight = lstm_mape / total_mape if total_mape > 0 else 0.5
    lstm_weight = sarima_mape / total_mape if total_mape > 0 else 0.5

    hybrid_fc = sarima_fc * sarima_weight + lstm_fc * lstm_weight

    results[group] = {
        'SARIMA': sarima_fc,
        'LSTM': lstm_fc,
        'Hybrid': hybrid_fc,
        'History': lstm_history,
        'CV_Metrics': cv_results[group]
    }

    # Plot combined results
    fig, ax = plt.subplots(figsize=(15, 6))
    ax.plot(series.index[-60:], series[-60:], label='Historical', color='blue', linewidth=2)
    ax.plot(hybrid_fc.index, hybrid_fc, label='Hybrid Forecast', color='red', linestyle='--', linewidth=2)
    ax.plot(sarima_fc.index, sarima_fc, label='SARIMA Forecast', color='green', alpha=0.5, linestyle=':')
    ax.plot(lstm_fc.index, lstm_fc, label='LSTM Forecast', color='purple', alpha=0.5, linestyle='-.')

    ax.set_title(f'{group} Consumption - Historical vs Forecast', pad=20)
    ax.set_xlabel('Year')
    ax.set_ylabel('Consumption (GWh)')
    ax.legend(loc='upper left')
    ax.xaxis.set_major_locator(YearLocator())
    ax.xaxis.set_major_formatter(DateFormatter('%Y'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Cross-validation results visualization (matplotlib)
metrics = ['MAE', 'RMSE', 'MAPE']
for metric in metrics:
    fig, ax = plt.subplots(figsize=(10, 6))
    x = np.arange(len(numeric_cols))
    bar_width = 0.35

    sarima_vals = [cv_results[group]['SARIMA'][metric] for group in numeric_cols]
    lstm_vals = [cv_results[group]['LSTM'][metric] for group in numeric_cols]

    ax.bar(x - bar_width/2, sarima_vals, bar_width, label=f'SARIMA {metric}', color='teal')
    ax.bar(x + bar_width/2, lstm_vals, bar_width, label=f'LSTM {metric}', color='coral')

    ax.set_xlabel('Consumption Group')
    ax.set_ylabel(metric)
    ax.set_title(f'Cross-Validation {metric} Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(numeric_cols, rotation=45)
    ax.legend()
    plt.tight_layout()
    plt.show()

```
